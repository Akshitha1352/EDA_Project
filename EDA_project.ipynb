# Exploratory Data Analysis of Fresher Hiring Trends in India


This project analyzes fresher hiring data to identify the factors that influence recruitment success and salary offers. Using Exploratory Data Analysis (EDA), the study examines the impact of CGPA, internships, certifications, projects, referrals, and skills on hiring outcomes, providing valuable insights into current recruitment trends in India.




# **Project Objective**

Analyze fresher hiring data to identify

Factors affecting hiring success.

Impact of CGPA, internships, projects, and certifications

 Salary trends

 Hiring stages distribution

 Most demanded skills

 Sector-wise hiring trends

**Business Problem Statement**

**Problem:**
Companies receive thousands of fresher applications. Understanding which factors contribute most to hiring success can help candidates improve their profiles and help recruiters optimize their hiring process.


**Goal:**
 Identify patterns in fresher recruitment and determine key attributes that influence selection and salary offers.

# IMPORT LIBRARIES

```python
# import requried libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOAD DATA

```python
df = pd.read_csv("/content/fresher_hiring_india_dataset.csv")

df.info()
```

# DATA CLENAING

Fill Interview Rounds

```python
df['interview_rounds'] = df['interview_rounds'].fillna(0)
```

Fill Salary

```python
df['offered_salary_inr'] = df['offered_salary_inr'].fillna(0)
```

Fill LinkedIn Premium

```python
df['linkedin_premium'] = df['linkedin_premium'].fillna('Unknown')
```

```python
df.info()
```

**Non-Visual Analyses using GroupBy, Pivot Table, and Crosstab**

1.Group By (Average CGPA by Hiring Stage (GroupBy))

```python
df.groupby('hiring_stage')['cgpa'].mean().sort_values(ascending=False)
```

# **Insights:**
**Candidates in advanced hiring stages generally have higher CGPA.**  
**CGPA plays an important role in shortlisting freshers.**

**Companies tend to prefer candidates with strong academic performance.**


**Internship Impact on Hiring Stage (Crosstab)**

```python
pd.crosstab( df['prior_internship'], df['hiring_stage'], margins=True )
```

# Insights:
**Candidates with internship experience progress further in the hiring process.**


**Internships improve practical knowledge and employability.**


**Recruiters view internship experience as a strong indicator of job readiness.**

**Salary by Sector and Degree (Pivot Table)**

```python
pd.pivot_table( df, values='offered_salary_inr', index='sector', columns='degree', aggfunc='mean' )
```

# Insights:
**Salary varies based on both educational qualification and industry sector.**

**Certain degrees command higher salaries in specific sectors.**

**This analysis helps identify the best degree-sector combinations for higher earnings.**

# Univariate Analysis
**Gender Distribution**

```python
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

**Degree Distribution**

```python
plt.figure(figsize=(10,5))
sns.countplot(y='degree', data=df)
plt.show()
```

**CGPA Distribution**

```python
sns.histplot(df['cgpa'], bins=20, kde=True)
plt.show()
```

# Bivariate Analysis
**CGPA vs Hiring Stage**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='hiring_stage', y='cgpa', data=df)
plt.xticks(rotation=45)
plt.show()
```

# Insight:

**Higher CGPA students are more likely to reach advanced hiring stages.**

**Certifications Impact vs Hiring stage**

```python
sns.boxplot(
    x='hiring_stage',
    y='certifications_count',
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

# Insight:

**Candidates with internships get shortlisted more frequently.**

**Certifications Impact vs Hiring Stage**

---



```python
sns.boxplot(
    x='hiring_stage',
    y='certifications_count',
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

# Insights:
**Candidates with multiple certifications are more likely to receive interview opportunities and progress further in the recruitment process.**

**Projects Impact vs Hiring stage**

```python
sns.boxplot(
    x='hiring_stage',
    y='projects_count',
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

# Salary Analysis
**Salary Distribution**

```python
salary_df = df[df['offered_salary_inr'] > 0]

sns.histplot(
    salary_df['offered_salary_inr'],
    bins=10
)

plt.show()
```

**Salary by Sector**

```python
plt.figure(figsize=(12,5))

sns.boxplot(
    x='sector',
    y='offered_salary_inr',
    data=salary_df
)

plt.xticks(rotation=90)
plt.show()
```

# Most Demanded Skills

```python
skills = df['top_skills'].str.split(',')

all_skills = []

for skill_list in skills:
    all_skills.extend(skill_list)

skills_df = pd.Series(all_skills)

skills_df.value_counts().head(15)
```

**Visualization:**

```python
skills_df.value_counts().head(15).plot(kind='barh')
plt.show()
```

# Insights:
Most demading skills are Deep learning, Bash Scripting, Hive, Linux, Communication, Django, Swift, JavaScript, Scala, Ruby.

# Top Hiring Companies

```python
top_companies = df['company_applied'].value_counts().head(10)

top_companies.plot(kind='bar')
plt.title("Top Hiring Companies")
plt.show()
```

# Correlation Analysis

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

# Insights:
**That academic performance, project experience, certifications, and recruitment progress collectively contribute to fresher hiring success and salary outcomes.**

# Conclusion

This Exploratory Data Analysis (EDA) project examined the key factors influencing fresher hiring outcomes in India. The analysis revealed that candidates with strong academic performance, internship experience, relevant certifications, and practical projects are more likely to progress through hiring stages and receive better salary offers.
